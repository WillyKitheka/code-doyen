```python
# fetch the file paths. That's the megajp_files.

import os

def get_csv_files(folder_path):
    csv_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.csv'):
                csv_files.append(os.path.join(root, file))
    return csv_files


```

```python
#function for creating a dictionary to load the csv files

import pandas as pd
import os

def load_csv_files(csv_files):
    megajp_sets = {}
    for file in csv_files:
        df = pd.read_csv(file)
        name = os.path.basename(file)
        name = os.path.splitext(name)[0]
        name = name.replace('-', '_')
        megajp_sets[name] = df
    return megajp_sets
```

```python
# same function same job scope! Except that this one has an exception for error printing...just in case(it's good practice)

def load_csv_files(csv_files):
    megajp_sets = {}
    for file in csv_files:
        try:
            df = pd.read_csv(file)
            name = os.path.basename(file)
            name = os.path.splitext(name)[0]
            name = name.replace('-', '_')
            megajp_sets[name] = df
        except pd.errors.ParserError as e:
            print(f"{file} could not be loaded due to error: {e}")
            continue
    return megajp_sets
```

```python
# conditional IF STATEMENT function. It will create 44 files as output based on the comparison between the megajp_sets and the 
# sample_space

def compare_columns(megajp_sets, sample_space):
    # Create an empty dictionary to store the results
    results = {}
    
    # Loop through the test sets
    for name, df in megajp_sets.items():
        # Check if the FTR column exists in df
        if 'FTR' in df.columns:
            # Compare the FTR column in df with each column in sample_space
            comparison = (df['FTR'].values[:, None] == sample_space).astype(int)
            comparison.columns = sample_space.columns
            # Add the comparison data frame to the results dictionary using the name of the test set as the key
            results[name] = comparison
        else:
            # If the FTR column does not exist, add an empty data frame to the results dictionary
            results[name] = pd.DataFrame(columns=sample_space.columns)
    
    return results
```

```python
# we create a function that takes a dictionary and 'score' as arguments and returns a new df with the filtered columns

def filter_columns(dfDict, score):

  # initialize a new data frame
  df_new = pd.DataFrame()

  # loop through the dictionary
  for i in dfDict:
    # select the row 'score' of the data frame
    row_values = dfDict[i].loc['score']
    # filter the columns of the data frame and retain only the columns that have a value of 8 in the row 'score'
    filtered_cols = row_values[row_values == score]
    # concatenate the filtered columns to the new data frame
    df_new = pd.concat([df_new, filtered_cols], axis = 1)
  # set the columns of the new data frame to the keys of the dictionary
  df_new.columns = dfDict.keys()

   # Keep all rows, whether duplicate or not
  df_new = df_new.drop_duplicates(keep=False)
  # return the new data frame
  return df_new
```

```python
import os
import pandas as pd

def search_column(paths, column_name):
    results = []
    for path in paths:
        df = pd.read_csv(path)
        if column_name in df.columns:
            parent_folder = os.path.basename(os.path.dirname(path))
            file_name = os.path.basename(path)
            results.append([parent_folder, file_name, path])
    return pd.DataFrame(results, columns=['Parent Folder', 'File Name', 'File Path'])
```

```python
import os
import pandas as pd

def search_column_vectorized(paths, column_name):
    results = []
    for path in paths:
        try:
            df = pd.read_csv(path, usecols=[column_name])
            parent_folder = os.path.basename(os.path.dirname(path))
            file_name = os.path.basename(path)
            results.append([parent_folder, file_name, path])
        except ValueError as e:
            pass
    return pd.DataFrame(results, columns=['Parent Folder', 'File Name', 'File Path'])
```

```python
import os
import dask.dataframe as dd

def search_column_by_dask(paths, column_name):
    results = []
    for path in paths:
        df = dd.read_csv(path, usecols=[column_name])
        if column_name in df.columns:
            parent_folder = os.path.basename(os.path.dirname(path))
            file_name = os.path.basename(path)
            results.append([parent_folder, file_name, path])
    return pd.DataFrame(results, columns=['Parent Folder', 'File Name', 'File Path'])
```
