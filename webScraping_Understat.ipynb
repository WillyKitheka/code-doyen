```python
#importing the essential libraries

import numpy as np #good with linear algebra

import pandas as pd #good with data processing, CSV I/O(e.g. pd.read_csv())

import requests

from bs4 import BeautifulSoup

```

```python
# in any webscraping project the first thing to do is research the website. Understand the layout and how 
# it works. That's fundamental. On the home page we notice that the site has data for 6 European Leagues.
# we also notice that the data available starts from season 2014/2015. It's link is represented as
# https://understat.com/league + name_of_league + year_start_of_the_season
# so we create variables with this data to be able to select any season or any league
# create urls for all seasons of all leagues

base_url = 'https://understat.com/league'

leagues = [ 'EPL', 'La_liga', 'Bundesliga', 'Serie_A', 'Ligue_1', 'RFPL']

seasons = ['2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']

```

```python
#starting with the latest data for EPL, beccause am an Arsenal fan

url = base_url + '/' + leagues[0]+ '/' + seasons[8]

results_understat = requests.get(url)

soup = BeautifulSoup(results_understat.content, 'lxml')

# Based on the structure of the web page, I found out that the data is in the JSON varible, under 
# 'script' tags

scripts = soup.find_all('script')
```

```python
# from inspecting the website we discovered that the data lives under a 'script' tag stored in a variable 
# called teamsData, after creating a soup of all the html tags it becomes just a string, so need to find 
# that text and extract JSON from it

#first, we import the library json{}

import json

string_with_json_obj = ''

# find the data for teams

for el in scripts:
  if 'teamsData' in el.text:
    string_with_json_obj = el.text.strip()

    print(string_with_json_obj) 

# strip unnecessary symbols and retain only JSON data

ind_start = string_with_json_obj.index("('")+2
ind_end = string_with_json_obj.index("')")
json_data = string_with_json_obj[ind_start:ind_end]

json_data = json_data.encode('utf8').decode('unicode_escape')

  
```

```python
# convert JSON data into a Python dictionary

data_understat = json.loads(json_data)
print(data_understat.keys()) 
```

```python
# get teams and their relevant ids and put them into separate dictionaries

teams = {}
for id in data_understat.keys():
  teams[id] = data_understat[id]['title']
```

```python
# although I can (but I will not) look for Arsenal's id. Here we just use Aston Villa id=71

aston_villa_data = []

for row in data_understat['71']['history']:
  aston_villa_data.append(list(row.values()))

df = pd.DataFrame(aston_villa_data, columns = columns)
df.head(5)
```

```python
# getting data for all teams

dataframes = {}

for id, team in teams.items():
  teams_data = []
  for row in data_understat[id]['history']:
    teams_data.append(list(row.values()))

    df = pd.DataFrame(teams_data, columns = columns)
    dataframes[team] = df
    print('Added data for {}.'.format(team))
```

```python
dataframes['Arsenal'].head(5)
```

```python
# manipulations to make the data appear as in does in the original source. For starters we notice that here such metrics as PPDA and OPPDA
# are represented as total amounts of 'att' read attacking and 'def' read defensive actions, but in the original table it is shown as 
# a coefficient. Let's fix that!

for team, df in dataframes.items():
  dataframes[team]['ppda_coef'] = dataframes[team]['ppda'].apply(lambda x: x['att']/x['def'] if x['def'] != 0 else 0)
  dataframes[team]['oppda_coef'] = dataframes[team]['ppda_allowed'].apply(lambda x:x['att']/x['def'] if x['def'] != 0 else 0)

# and now let's check how our new dataframes look based on the aston_villa df

dataframes['Aston Villa'].head(5)
```

```python
# we have all our numbers, but for every single match. What we need is the totals for the team. So we find the columns and sum them up.
# for this we go back to the original table at understat.com and find all the metrics that are summed up. We also discover that PPDA and
# OPPDA are means in the end

cols_to_sum = ['xG', 'xGA', 'npxG', 'npxGA', 'deep', 'deep_allowed', 'scored', 'missed', 'xpts', 'wins', 'draws', 'loses', 'pts', 'npxGD']

cols_to_mean = ['ppda_coef', 'oppda_coef']
```

```python
# with that we are ready to calculate our totals adn means. For this we loop through the dictionary and call .sum() and .mean() DataFrame 
# methods that return Series, that's why we add .transpose() to those calls. We then put these new DataFrames into list and after that we
# concat them into a new DataFrame 'full_stat'

frames = []

for team, df in dataframes.items():
  sum_data = pd.DataFrame(df[cols_to_sum].sum()).transpose()
  mean_data =  pd.DataFrame(df[cols_to_mean].mean()).transpose()
  final_df = sum_data.join(mean_data)
  final_df['team'] = team
  final_df['matches'] = len(df)
  frames.append(final_df)

  full_stat = pd.concat(frames)
```

```python
from pickle import FALSE
# next we reorder columns for better readability, sort rows based on points, reset index and add column 'position'

full_stat = full_stat[['team', 'matches', 'wins', 'draws', 'loses', 'scored', 'missed', 'pts', 'xG', 'npxG', 'xGA', 'npxGA', 'npxGD', 'ppda_coef', 'oppda_coef', 'deep', 'deep_allowed', 'xpts']]
full_stat.sort_values('pts', ascending = False, inplace = True)
full_stat.reset_index(inplace = True, drop = True)
full_stat['position'] = range(1, len(full_stat)+1)
```

```python
# in the orginal table we also have values of differences between expected metrics and real. Let's add these too!

full_stat['xG_diff'] = full_stat['xG'] - full_stat['scored']
full_stat['xGA_diff'] = full_stat['xGA'] - full_stat['missed']
full_stat['xpts_diff'] = full_stat['xpts'] - full_stat['pts']
```

```python
# where appropriate, converting floats to integers

cols_to_int = ['wins', 'draws', 'loses', 'scored', 'missed', 'pts', 'deep', 'deep_allowed']
full_stat[cols_to_int] = full_stat[cols_to_int].astype(int)
```

```python
# prettifying final output and view of a DataFrame

col_order = ['position', 'team', 'matches', 'wins', 'draws', 'loses', 'scored', 'missed', 'pts', 'xG', 'xG_diff', 'npxG', 'xGA', 'xGA_diff', 'npxGA', 'npxGD', 'ppda_coef', 'oppda_coef', 'deep', 'deep_allowed', 'xpts', 'xpts_diff']
full_stat = full_stat[col_order]
pd.options.display.float_format = '{:,.2f}'.format
full_stat.head(10)
```

```python
full_data = dict()
for league in leagues:

  season_data = dict()
  for season in seasons:
    url = base_url + '/' + leagues[0]+ '/' + seasons[8]

    results_understat = requests.get(url)

    soup = BeautifulSoup(results_understat.content, 'lxml')

# Based on the structure of the web page, I found out that the data is in the JSON varible, under 
# 'script' tags

    scripts = soup.find_all('script')
    string_with_json_obj = ''

# find the data for teams

    for el in scripts:
        if 'teamsData' in el.text:
          string_with_json_obj = el.text.strip()
    
# strip unnecessary symbols and retain only JSON data

    ind_start = string_with_json_obj.index("('")+2
    ind_end = string_with_json_obj.index("')")
    json_data = string_with_json_obj[ind_start:ind_end]
    json_data = json_data.encode('utf8').decode('unicode_escape')
    data_understat = json.loads(json_data)

    # get teams and their relevant ids and put them into separate dictionaries

    teams = {}
    for id in data_understat.keys():
      teams[id] = data_understat[id]['title']
    
    # EDA to get a feel of how the JSON is structured
    # column names are all the same; so wew just use first element

    columns = []

    # check the sample values per each column

    values = []
    for id in data_understat.keys():
      columns = list(data_understat[id]['history'][0].keys())
      values = list(data_understat[id]['history'][0].values())
      break
    
    # getting data for all teams

    dataframes = {}

    for id, team in teams.items():
      teams_data = []
      for row in data_understat[id]['history']:
        teams_data.append(list(row.values()))

        df = pd.DataFrame(teams_data, columns = columns)
        dataframes[team] = df

# manipulations to make the data appear as in does in the original source. For starters we notice that here such metrics as PPDA and OPPDA
# are represented as total amounts of 'att' read attacking and 'def' read defensive actions, but in the original table it is shown as 
# a coefficient. Let's fix that!

    for team, df in dataframes.items():
      dataframes[team]['ppda_coef'] = dataframes[team]['ppda'].apply(lambda x: x['att']/x['def'] if x['def'] != 0 else 0)
      dataframes[team]['oppda_coef'] = dataframes[team]['ppda_allowed'].apply(lambda x:x['att']/x['def'] if x['def'] != 0 else 0)

    cols_to_sum = ['xG', 'xGA', 'npxG', 'npxGA', 'deep', 'deep_allowed', 'scored', 'missed', 'xpts', 'wins', 'draws', 'loses', 'pts', 'npxGD']
    cols_to_mean = ['ppda_coef', 'oppda_coef']

    frames = []

    for team, df in dataframes.items():
      sum_data = pd.DataFrame(df[cols_to_sum].sum()).transpose()
      mean_data =  pd.DataFrame(df[cols_to_mean].mean()).transpose()
      final_df = sum_data.join(mean_data)
      final_df['team'] = team
      final_df['matches'] = len(df)
      frames.append(final_df)

    full_stat = pd.concat(frames)

    full_stat = full_stat[['team', 'matches', 'wins', 'draws', 'loses', 'scored', 'missed', 'pts', 'xG', 'npxG', 'xGA', 'npxGA', 'npxGD', 'ppda_coef', 'oppda_coef', 'deep', 'deep_allowed', 'xpts']]
    full_stat.sort_values('pts', ascending = False, inplace = True)
    full_stat.reset_index(inplace = True, drop = True)
    full_stat['position'] = range(1, len(full_stat)+1)

    full_stat['xG_diff'] = full_stat['xG'] - full_stat['scored']
    full_stat['xGA_diff'] = full_stat['xGA'] - full_stat['missed']
    full_stat['xpts_diff'] = full_stat['xpts'] - full_stat['pts']

    cols_to_int = ['wins', 'draws', 'loses', 'scored', 'missed', 'pts', 'deep', 'deep_allowed']
    full_stat[cols_to_int] = full_stat[cols_to_int].astype(int)
    col_order = ['position', 'team', 'matches', 'wins', 'draws', 'loses', 'scored', 'missed', 'pts', 'xG', 'xG_diff', 'npxG', 'xGA', 'xGA_diff', 'npxGA', 'npxGD', 'ppda_coef', 'oppda_coef', 'deep', 'deep_allowed', 'xpts', 'xpts_diff']
    full_stat = full_stat[col_order]
    full_stat = full_stat.set_index('position')
    # print (full_stat.head(20))

    season_data[season] = full_stat

  df_season = pd.concat(season_data)
  full_data[league] = df_season

data = pd.concat(season_data)
data.head()



```
