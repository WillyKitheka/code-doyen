```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# create table A
table_A = pd.DataFrame(columns=['date', 'league', 'home_team', 'away_team', 'home_score', 'away_score', 'classResult'])
table_A['date'] = pd.date_range(start='2022-01-01', end='2022-01-05')
table_A['league'] = ['Premier League', 'Serie A', 'La Liga', 'Bundesliga', 'Ligue 1']
table_A['home_team'] = ['Arsenal', 'AC Milan', 'Barcelona', 'Bayern Munich', 'Paris Saint-Germain']
table_A['away_team'] = ['Manchester United', 'Juventus', 'Real Madrid', 'Borussia Dortmund', 'Olympique Marseille']
table_A['home_score'] = np.random.randint(0, 4, size=len(table_A))
table_A['away_score'] = np.random.randint(0, 4, size=len(table_A))
table_A['classResult'] = np.where(table_A['home_score'] > table_A['away_score'], 'H', np.where(table_A['home_score'] < table_A['away_score'], 'A', 'D'))

# create table B
table_B = pd.DataFrame(columns=['date', 'league', 'home_team', 'away_team', 'xG', 'xGA', 'xGD', 'oppda_coef', 'ppda_coef'])
table_B['date'] = pd.date_range(start='2022-01-01', end='2022-01-05')
table_B['league'] = ['Premier League', 'Serie A', 'La Liga', 'Bundesliga', 'Ligue 1']
table_B['home_team'] = ['Arsenal', 'AC Milan', 'Barcelona', 'Bayern Munich', 'Paris Saint-Germain']
table_B['away_team'] = ['Manchester United', 'Juventus', 'Real Madrid', 'Borussia Dortmund', 'Olympique Marseille']
table_B['xG'] = np.round(np.random.uniform(0.5, 3, size=len(table_B)), 2)
table_B['xGA'] = np.round(np.random.uniform(0.5, 3, size=len(table_B)), 2)
table_B['xGD'] = np.round(table_B['xG'] - table_B['xGA'], 2)
table_B['oppda_coef'] = np.round(np.random.uniform(5, 25, size=len(table_B)), 2)
table_B['ppda_coef'] = np.round(np.random.uniform(5, 25, size=len(table_B)), 2)

# set same dates in both tables
table_A.loc[[1, 2, 3], 'date'] = table_B.loc[[1, 2, 3], 'date']

# print tables
print('Table A:')
print(table_A)
print('\nTable B:')
print(table_B)
```

```python
import pandas as pd
from datetime import datetime, timedelta
import random

# Define start and end dates for the data
start_date = datetime(2022, 1, 1)
end_date = datetime(2022, 1, 7)

# Generate data for table A
data_a = {
    'date': [start_date + timedelta(days=i) for i in range(5)],
    'league': ['League A']*5,
    'home_team': ['Team {}'.format(i) for i in range(1, 6)],
    'away_team': ['Team {}'.format(i) for i in range(2, 7)],
    'home_score': [random.randint(0, 5) for _ in range(5)],
    'away_score': [random.randint(0, 5) for _ in range(5)]
}
table_a = pd.DataFrame(data_a)

# Generate data for table B
data_b = {
    'date': [start_date + timedelta(days=i, hours=12) for i in range(5)],
    'league': ['League A']*5,
    'home_team': ['Team {}'.format(i) for i in range(1, 6)],
    'away_team': ['Team {}'.format(i) for i in range(2, 7)],
    'xG': [random.uniform(0, 2) for _ in range(5)],
    'xGA': [random.uniform(0, 2) for _ in range(5)],
    'xGD': [random.uniform(-1, 1) for _ in range(5)],
    'oppda_coef': [random.uniform(10, 20) for _ in range(5)],
    'ppda_coef': [random.uniform(0, 5) for _ in range(5)]
}
table_b = pd.DataFrame(data_b)

# create an empty list to store the rows of table_c
table_c_rows = []

# loop through each row in table_a
for index, row_a in table_a.iterrows():
    
    # get the date, home team and away team from the current row of table_a
    date = row_a['date']
    home_team = row_a['home_team']
    away_team = row_a['away_team']
    
    # filter table_b based on date, home team and away team
    filtered_b = table_b.loc[(table_b['date'] == date) & (table_b['home_team'] == home_team) & (table_b['away_team'] == away_team)]
    
    # check if there is a match in table_b
    if len(filtered_b) > 0:
        # merge the row from table_a with the matching row from table_b
        merged_row = pd.concat([row_a, filtered_b.iloc[0]])
        table_c_rows.append(merged_row)

# create table_c from the list of rows
table_c = pd.DataFrame(table_c_rows)

# Print tables A, B and C
print('Table A:')
print(table_a)
print('\nTable B:')
print(table_b)
print('\nTable C:')
print(table_c)
```
